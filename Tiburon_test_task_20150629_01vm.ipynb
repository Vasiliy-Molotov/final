### Методы опознавания недобросовестной анкеты

Под недобросовестными анкетами будем понимать те, которые заполнены безразлично к содержанию вопросов. Это те анкеты, в которых участник не задумывался над вопросами.

    Это упрощенное определение. В более общем случае участник может заполнять анкету частично добросовестно, частично - недобросовестно. Тогда можно говорить об уровне качества заполнения по частям анкеты, что применимо для длинных опросов. Также он может приспособиться к малоосмысленному прохождению анкеты, при этом давая ответы, которые успешно принимаются алгоритмом проверки качества).

    Такие анкеты могут быть опознаны по следующим признакам:
    применимым в данном случае:
    0) противоречия между ответами на связанные по смыслу вопросы
    1) аномальное время заполнения
    2) стериотипное заполнение (все ответы "полностью согласен" и тп.)
    
    дополнительно применимым в общем случае:
    3) невозможные\неверные ответы на вопросы и вопросы-ловушки ("выберите колонку №2" и тп.)
    4) много ответов не имеющих содержательной информации ("затрудняюсь ответить"/"не знаю"/"прочее" и тп.)
    5) противоречие между ответами и профилем участника (по данным регистрационной анкеты \ профилю в соц. сети )
    6) противоречие между ответами в данном опросе и предыдущих опросах
    7) противоречие по геолокации по IP (заполнен другим человеком по данному аккаунту) \ заходы по одной уникальной ссылке на опрос с разных IP

### Противоречивые вопросы

Следующие пары вопросов подразумевают, что при осмысленном ответе участник даст либо одинаковые по полярности ответы (оба раза согласен, либо оба раза не согласен), либо разнонаправленные (согласен и не согласен, либо не согласен и согласен)

    Сонаправленные по уровню согласия вопросы ( id опроса 1 / id опроса 2 / ... )
     0 / 5
     1 / 8
     3 / 12    
     7 / 13  
     10 / 14 (можно добавить 9)
     
    Разнонаправленные
    1 / 10
    8 / 10
    1 / 14
    8 / 14
    
    Итого имеем 9 пар связанных вопросов, по которым можно проверить достоверность анкеты.

Рассмотрим вероятность дать противоречивые ответы на пару связанных вопросов при заполнении анкеты наугад. Будем считать, что заполняющий выбирает случайную оценку на каждый вопрос
    
    (при этом следует учесть, что человек не является хорошим генератором случайных чисел, и при достаточном количестве заполненных наугад опросов от конкретного респондента можно выделить его паттерн поведения. Собрав достаточный набор наугад заполненных опросов можно составить обучающую базу для упрощения задачи идентификации недобросовестно заполненных анкет по сравнению на уровне паттернов)
    
    В паре сонаправленных вопросов противоречивых ответами будем считать следующие ответы (ответ на вопрос 1 / ответ на вопрос 2)
    1) "полностью согласна" / "скорее не согласна"
    2) "полностью согласна" / "полностью не согласна"
    3) "скорее согласна" / "скорее не согласна"
    4) "скорее согласна" / "полностью не согласна"
    И еще 4 такие же комбинации, где эти же ответы даны в обратном порядке (ответ на вопрос 2 / ответ на вопрос 1)
    Всего 8 вариантов противоречивых ответов

    Всего 25 возможных комбинаций ответов на эту пару вопросов
    Поэтому вероятность дать противоречивый ответ наугад составляет 8/25 = 0.32
    
    

Принимая во внимание, что среди 9 пар связанных вопросов есть пересечения (например, вопрос 1 входит в 3 пары), сделаем упрощение, что пары вопросов не связаны и аппроксимируем распределение вероястности дать определенное количество противоречивых ответов распределением Бернулли (количество экспериментов = 9, вероятность положительного исхода = 0.32)

```python
from scipy.stats import binom
rv = binom(9, 0.32, loc=0)
sp = 1
print "вероятность p дать ровно i противоречивых ответов, вероятность sp дать как минимум i противоречивых ответов на 9 пар связанных вопросов, выбирая ответы наугад" 
for i in range(10):
    p = rv.pmf(i)
    print "i = %d | p = %f | sp = %f" % (i, p, sp)
    sp-= p
```

Отсюда следует, что большинство "выберателей наугад" с 97% вероятностью выберут противоречащие ответы как минимум единожды, с 83% - дважды, в среднем они выбирт 2-4 противоречивых ответа.

### Время заполнения

Для каждого опроса целесообразно строить его индивидуальный график распределения анкет по времени, затраченному на ее заполненения. При этом для каждого опроса имеются следующие 3 определяющих показателя:

    1) t_1 минимальное время, необходимое для осмысленного заполнения (экспертная оценка)
    2) t_2 среднее время, затраченное всеми участниками на заполнение анкеты (как вариант - без учета "выбросов")
    3) t_3 среднее время для неосмысленного заполненения (экпериментальная оценка)

По графику распределения исходя из его формы и вышеперечисленных параметров выбирается допустимый диапазон "реалистичного осмысленного времени заполнения" [t_min, t_max].

Параметры диапазона имеет смысл выбирать исходя из его конкретной формы распределения, а не руководствуясь заранее заданной формулой, в силу следующих причин:

    1) Если, например, все анкеты были заполнены на скорость и не являются достоверными, то они все отсеиваются по показателю t_1, при этом показатель t_2 меньше, чем t_1, и допустимый диапазон осмысленного заполненеия на его основе построен быть не может. Из этого следует необходимость задания t_1
    
    2) В зависимости от сложности, длинны опроса, вариантов выбора и тп. диапазон допустимого времени заполнения может меняться 
    
    3) Скорее всего, большинство анкет будет сосредоточено в 2 областях: в районе t_3 - недобросовестные анкеты, и в районе t_2 +- delta. И эти области будут чутко разделимы.
    
Все анкеты маркируются пометками - "реалистичное время", "нереалистичное время"

### Стереотипное заполнение

База стереотипных заполнений может пополняться и зависеть от формата анкеты

    Здесь зададим 2 типа стереотипных ответов
    1) все ответы одинаковы (составляют единую колонку)
    2) зиг-заг - ответы в анкете можно разделить на блоки по 5 последовательных вопросов, где ответы для каждого блока составляют его диагональ

### Итоговый алгоритм

    Возможны разные подходы:
    1) детерминированный - каждая анкета либо "добросовестная", либо нет
    2) для каждой анкеты есть вероятность ее "добросовестности"

    Рассмотрим 1 вариант:

    Дано:
    <№ интервью><Время заполнения в сек><Оценка 1><Оценка 2>…<Оценка N>
    ...

    0) Задать параметры качества
    0.1) Экспертно определить t_1, t_3 (зависят от формата опроса, где и как проводится)
    0.2) Задать уровень допустимого количества противоречивых ответов для брака анкеты n_min = 2 (уровень определенности ~83%)

    1) Для каждого <№ интервью> 
    1.1) расчитать n - количество противоречивых ответов
    1.2) расчитать s - (s = 1, если стереотипное заполнение, s = 0 иначе)
    
    2) построить график распределения времени заполнения анкеты, проанализировать его и расчитать t_2
    2.1) задать диапазон [t_min, t_max]
    2.2) промаркировать каждую анкету t (t = 0 для допустимого времени, t=1 для недопустимого)
    
    3) Имея комбинацию вышеперечисленных факторов можно построить нужный по уровню детализации алгоритм
    Здесь пусть будет отсев по каждому признаку:
    
    если n >= n_min
    или
    если t = 1
    или
    если s = 1
    то
    анкета "недобросовестная"

```python
rng = pd.DataFrame({'start_time':pd.date_range('29/6/2015', periods=72, freq='min')})
rng['start_time'] = rng['start_time'] + (500*(np.random.rand(len(rng),1))).astype(int).astype('m8[s]')[:,0]
rng['duration'] = np.exp(2*(np.random.rand(len(rng),1))+4).astype(int) #lognormal distribution
rng['end_time'] = rng['duration'].astype('m8[s]')+rng['start_time']
```

```python
df1 = pd.DataFrame({'time': rng['start_time'], 'type':'start'})
df2 = pd.DataFrame({'time': rng['end_time'], 'type':'end'})
df3 = pd.concat([df1,df2])
df3 = df3.sort(['time'],ascending=[1]).reset_index(drop=1).copy()
df3['n_calls_atm']=0
```

```python
df3.ix[0,'n_calls_atm'] = 1
for i in df3.index[1:]:
    df3.ix[i,'n_calls_atm'] = df3.ix[i-1,'n_calls_atm']
    if df3.ix[i,'type'] == 'start':
        df3.ix[i,'n_calls_atm']+=1
    else:
        df3.ix[i,'n_calls_atm']-=1
```

```python
ax = plt.plot(df3['time'],df3['n_calls_atm'])
pylab.savefig("call-center", bbox_inches='tight')
plt.show()
```
